```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
pd.set_option("display.max_columns", None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 200)
```

```python

BASE_DIR = '/home/thanuja/Dropbox/coursera/Milestone1/data/'
```

```python
#getting vote share by zip
vote_file = pd.read_csv(BASE_DIR + "countypres_2020_pivoted.csv")
vote_file['county_fips'] = vote_file['county_fips'].dropna().astype(int)
zip_county = pd.read_csv(BASE_DIR + "ZIP_COUNTY_092021.csv").dropna()
zip_county['county_fips'] = zip_county['COUNTY'].astype(int)

votes_by_county = vote_file.merge(zip_county, on='county_fips')

votes_by_county = votes_by_county[['ZIP', 'COUNTY', 'USPS_ZIP_PREF_CITY', 'USPS_ZIP_PREF_STATE', 'DEMOCRAT','REPUBLICAN','GREEN','LIBERTARIAN','OTHER']]
for col in ['DEMOCRAT','REPUBLICAN','GREEN','LIBERTARIAN','OTHER']:
    votes_by_county[col] = votes_by_county[col].astype(float)

votes_by_zip = votes_by_county.groupby('ZIP').agg(
    DEMOCRAT=pd.NamedAgg(column='DEMOCRAT', aggfunc=np.mean),
    REPUBLICAN=pd.NamedAgg(column='REPUBLICAN', aggfunc=np.mean),
    GREEN=pd.NamedAgg(column='GREEN', aggfunc=np.mean),
    LIBERTARIAN=pd.NamedAgg(column='LIBERTARIAN', aggfunc=np.mean),
    OTHER=pd.NamedAgg(column='OTHER', aggfunc=np.mean)
)

votes_by_zip.to_csv(BASE_DIR + 'data_processing/vote_share/votes_by_zip.csv')
votes_by_zip
```
